```python
import pandas as pd

filename = "Donnees_projet_fil_rouge.csv"
df = pd.read_csv(filename, header=0, sep=';')  # essayez d'abord avec point-virgule

print(f"Dataset Shape: {df.shape}")

print(df.columns)
```

```python
print(df.head())
```

```python
#graphique des repartitions des differentes colonnes
import matplotlib.pyplot as plt 
import seaborn as sns

#colorier les graphiques pour séprarer les classes céréales et non céréales

fig, ax = plt.subplots(figsize=(10, 6))


for column in df.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column, palette='Set2')
    plt.title(f'Repartition of {column} by Cereal Class')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

#graphique des repartitions des differentes colonnes numeriques
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=column, kde=True, palette='Set2', bins=30)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()  


```
